```python
import pandas as pd
import ast

def safe_eval(x):
    try:
        return ast.literal_eval(x)
    except:
        return x

def expand_dataframe(df):
    # First, ensure all list-like columns are actually lists
    list_columns = ['original_facts', 'generated_facts', 'most_relevant_evidence', 'most_similar_facts', 'factuality_scores', 'similarity_scores']
    for col in list_columns:
        df[col] = df[col].apply(safe_eval)

    rows = []
    for idx, row in df.iterrows():
        original_text = row['original_text']
        generated_text = row['generated_text']
        
        # Determine the maximum length of all lists
        max_len = max(len(row[col]) for col in list_columns)
        
        for i in range(max_len):
            new_row = {
                'ad_idx': idx,
                'original_text': original_text,
                'generated_text': generated_text,
                'original_fact': row['original_facts'][i] if i < len(row['original_facts']) else None,
                'generated_fact': row['generated_facts'][i] if i < len(row['generated_facts']) else None,
                'most_relevant_evidence': row['most_relevant_evidence'][i] if i < len(row['most_relevant_evidence']) else None,
                'most_similar_fact': row['most_similar_facts'][i] if i < len(row['most_similar_facts']) else None,
                'factuality_score': row['factuality_scores'][i] if i < len(row['factuality_scores']) else None,
                'similarity_score': row['similarity_scores'][i] if i < len(row['similarity_scores']) else None
            }
            rows.append(new_row)
    
    return pd.DataFrame(rows)

# Assuming your DataFrame is named factuality_df
expanded_df = expand_dataframe(factuality_df)
expanded_df = expanded_df.drop(columns=['original_fact'])

# Display the first few rows of the new DataFrame
expanded_df.head(5)
expanded_df.to_csv('/dccstor/autofair/busekorkmaz/factual-bias-mitigation/data/output_files/expanded_df.csv', index=False)
```
